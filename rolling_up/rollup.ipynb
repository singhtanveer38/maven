```python
import pandas as pd
```

```python
df = pd.read_csv("coffee_shop_sales.csv", parse_dates=["date"])
df
```

```python
df.info()
```

```python
df["month"] = df["date"].dt.month
df
```

```python
grouped_df = df.groupby(by=["month", "store"]).aggregate({"sales": sum})
grouped_df
```

```python
grouped_df = grouped_df.reset_index()
grouped_df = grouped_df.sort_values(by=["store"])
grouped_df
```

```python
sales_diff = grouped_df[["store", "sales"]].groupby(by=["store"]).diff()
sales_diff
```

```python
rollup = pd.concat(objs=[grouped_df, sales_diff], axis=1)
rollup
```

```python
rollup = rollup.sort_values(by=["month"])
rollup
```

```python
def number_to_month(month):
    month_name = {
        1: "Jan",
        2: "Feb",
        3: "Mar",
        4: "Apr",
        5: "May",
        6: "Jun",
        7: "Jul",
        8: "Aug",
        9: "Sep",
        10: "Oct",
        11: "Nov",
        12: "Dec"}

    return month_name[month]
```

```python
rollup["month"] = rollup["month"].apply(lambda x: number_to_month(x))
rollup
```
