```python
import pandas as pd
import json
```

```python
df = pd.read_csv("sales_orders.csv", )
df.head()
```

```python
df["line_items"][0]
```

```python
df["line_items"] = df["line_items"].apply(json.loads)
df
```

```python
df = df.explode(column="line_items")
df
```

```python
line_items = pd.json_normalize(df["line_items"])
line_items
```

```python
orders = pd.concat([df.reset_index(), line_items], axis=1)
orders
```

```python
orders = orders.drop(labels=["index", "line_items"], axis=1)
orders
```

```python
orders["product.product_name"] = orders["product.product_name"].str.replace(",", " ")
```

```python
orders.to_csv("orders.csv", index=False)
```

```python
orders["total_cost"] = orders["product.product_price"] * orders["quantity"]
```

```python
orders["fulfillment"].value_counts()
```

```python
total_price = round(orders[orders["fulfillment"] == "Online"]["total_cost"].sum())
```
